In the case of creating neural networks, the sample which is gonna be created here is the one **where it learns the relationship between two numbers**. For that lets give out set of Xs and Ys, then it should be able to figure out the relationship b/w them.
This is obviously a different paradigm than what we might be used to, so lets step through piece by piece.

**Imports**
importing Tensorflow and calling it tf for easo of use.

importing numpy for representing the data as lists easily and quickly.

Keras is a Framework used for definig a Neural Network as a set of sequential layers. So importing Keras.

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

**Definie and Compile Neural Network**

Next we'll create the simplest possible Neural Network.
It has **onle Layer and one neuron and the input shape to it is just one value**.

```
model = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
```

Now we compile our Neural Network, When we do so we have to specify 2 functions:


1.   Loss
2.   optimizer

If you've seen lots of math for machine learning, here's where it's usually used, but in this case it's nicely encapsulated in functions for you. But what happens here -- let's explain...

We know that in our function, the relationship between the number is y = 3x+1

When the computer is trying to learn that, it makes a guess, maybe y = 10x+10.
The **LOSS function** compares the guessed answer with the known answer and measures how well or badly it did and the difference in error will be shown.

It then uses the Optimizer function to make another guess. Based on how the LOSS function went, the **OPTIMIZER** will try to minimize the loss. At that point maybe it will come up with somethinglike y = 5x+5, whis is still bad but compared to the previous guess, the loss is much lower.

It will repeat thi process to a number of **EPOCHS** (EPOCHS is where the neuralnetwork learns).

Here we are using **"MEAN SQUARED ERROR"** as the loss function and **"STOCHASTIC GRADIENT DESCENT"** fro the optimizer.
There are lots of other Optimizers and Loss gunctions as well.


```
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

**Providing the Data**

Next up we'll feed in some data. In this case we are taking 6 xs and 6ys. You can see that the relationship between these is that y=3X+1, so where x = -1, y=-3 etc. etc.

The values are specified as an np array provided by the Python library called Numpy

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype = float)
```

**Training the Neural Network**

The process of training the Neural Network, where it learns the relationship b/w xs and ys is in the **model.fit** call.

As mentioned above, this is where the Neural Network go in loop, ie, based on the back propagation.

It will do the EPOCHS as the number of epochs we specify to reach the **global minima**.

```
model.fit(xs, ys, epochs = 500)
```

Now the model has been trained to lern the relationship between X and Y,and as you can see the loss is the lowest when it is at the 500th epoch.

we can use the **model.predict** method to figure out a Y for a previously unknown X.

```
print(model.predict([10.0]))
```

You might have thought 31, right? But it ended up being a little over. Why do you think that is?

Remember that neural networks deal with probabilities, so given the data that we fed the NN with, it calculated that there is a very high probability that the relationship between X and Y is Y=3X+1, but with only 6 data points we can't know for sure. As a result, the result for 10 is very close to 31, but not necessarily 31.

As you work with neural networks, you'll see this pattern recurring. You will almost always deal with probabilities, not certainties, and will do a little bit of coding to figure out what the result is based on the probabilities, particularly when it comes to classification
